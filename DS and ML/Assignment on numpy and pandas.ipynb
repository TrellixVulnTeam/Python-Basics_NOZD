```python
df=pd.read_csv('olympics.csv',index_col=0,skiprows=1)
for col in df.columns:
    if col[:2]=='01':
        df.rename(columns={col:'Gold'+col[4:]},inplace=True)
    if col[:2]=='02':
        df.rename(columns={col:'Silver'+col[4:]},inplace=True)
    if col[:2]=='03':
        df.rename(columns={col:'Bronze'+col[4:]},inplace=True)
    if col[:1]=='№':
        df.rename(columns={col:'#'+col[1:]},inplace=True)
names_ids=df.index.str.split('\s\(')   #split the index by '('
df.index=names_ids.str[0]           # the [0] element is the country name (new index)
df['ID']=names_ids.str[1].str[:3]   #the [1] element is the abbreviation or ID (take first 3 character from that)
df=df.drop('Totals') 
df.head()
```

```python
def answer_zero():
    return df.iloc[0]
answer_zero()
```

```python
def answer_one():
    return df[df['Gold']==df['Gold'].max()].index[0]
answer_one()
```

```python
def answer_two():
    df['diff']=abs(df['Gold']-df['Gold.1']) # create new row containing abs value of gold differnce
    return df[df['diff']==df['diff'].max()].index[0] # index/name from the element with the largest difference
answer_two()
```

```python
def answer_three():
    sub_df=df[(df['Gold']>0) & (df['Gold.1']>0)].copy() # create sub dataframe of only elements with greater than 0 gold and gold.1 
    sub_df['rel']=sub_df['diff']/sub_df['Gold.2']# calculate diff between gold medals relative to total medal count
    return sub_df[sub_df['rel']==max(sub_df['rel'])].index[0]
answer_three()
```

```python
def answer_four():
    df['points']=df['Gold.2']*3 +df['Silver.2']*2+df['Bronze.2']*1
    return df['points']
answer_four()
```
