```python
labels=['python','c++','ruby','java','R']
sizes=[250,130,200,210,100]
colors=['yellow','yellowgreen','pink','blue','red']
exp=[0,0,0,0,0.1]
plt.pie(sizes,explode=exp,labels=labels,colors=colors,startangle=90)
plt.legend()
```

```python
language=['python','c++','ruby','java','R']
sizes=[250,130,200,210,100]
colors=['yellow','yellowgreen','pink','blue','red']
exp=[0,0.1,0,0,0.1]
plt.pie(sizes,explode=exp,labels=language,colors=colors,startangle=90,autopct='%.2f%%',shadow=True)
plt.legend()
```

```python
col=['red','green','blue']
siz=np.array([54,22,24])
val=['veg','non','not']
exp=[0,0,0]
def calc(values):
    a=np.round(values/100*siz.sum())
    return a
plt.pie(siz,labels=val,colors=col,explode=exp,startangle=90,autopct=calc)
mcircle=plt.Circle((0,0),.4,color='white')
p=plt.gcf()               #gcf= graphical configuration
p.gca().add_artist(mcircle)
plt.show()
```
